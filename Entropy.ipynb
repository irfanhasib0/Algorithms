```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
ents=[]
for i in range(1,10):
        x=i/10
        a=x;b=1-x
        ents+=[-a*np.log2(a)]
plt.plot(ents)
```

```python
ents=[]
for i in range(1,10):
        x=i/10
        a=x;b=1-x
        ents+=[-a*np.log2(a)-b*np.log2(b)]
plt.plot(ents)
```
